```python
def test(source,destination):
    locator = Nominatim(user_agent="myGeocoder")
    location = locator.geocode(source)
    sx=location.latitude
    sy = location.longitude
#     sx = -74.010004
#     sy = 40.630113
    print(sx,sy)
    loca = locator.geocode(destination)
    ex = loca.latitude
    ey = loca.longitude
#     ex = -74.0107
#     ey = 40.629646
    print(ex,ey)
    orig_node = ox.nearest_nodes(graph, sy,sx) # bug detected: order should be x,y but it expects y,x
    dest_node = ox.nearest_nodes(graph, ey,ex)
    shortest_route = nx.shortest_path(graph,
                                  orig_node,
                                  dest_node,
                                  weight='time')
    return (shortest_route) #@ had to convert to str for frontend
```

```python
path = test('Leif Ericson Park','Harmony Triangle')
```
